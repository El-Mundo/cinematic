```python
import plotly.express as px
import pandas as pd
```

```python
tarpath = "./region_probabilities.csv"

tiile = "Probabilities of Topics in Regions"

df = pd.read_csv(tarpath)
df = df.sort_values(by="Region")

fig = px.scatter(df, x="Region", y="Probability", color="Topic", hover_name="Topic", hover_data=["Probability","Region","Topic"], title=tiile)
fig.show()
```
