# Ultimate Gallery Thumbnails

This notebookd will save all extracted frames analysed from the videos (477479 images) to CV/thumbnails/cache

```python
import cv2
import glob
from datetime import datetime
import os
import json

TIME_INTERVAL = 10.0 # second(s)
SCALING_FACTOR = 0.3
REGION = "Beijing"

with open("../metadata-all.json") as f:
    films = json.load(f)["films"]
    
filmTitleToObj = {}

for film in films:
    special_name = False
    fKey = ""
    title = film["title"]
    if film["title"] == "母亲":
        fKey = "mu qin"
        special_name = True
    elif film["title"] == "欢天喜地":
        fKey = "huan tian xi di (1959)"
        special_name = True
    elif film["title"] == "春雷":
        fKey = "chun lei (1958)"
        special_name = True
    elif film["title"] == "穆桂英挂帅":
        fKey = "mu gui ying gua shuai (1958)"
        special_name = True
    elif film["title"] == "大李、小李和老李":
        title = "大李小李和老李"
    elif film["title"] == "“特快”列车":
        title = "特快列车"
    elif film["title"] == "打铜锣、补锅":
        title = "打铜锣补锅"
    elif film["title"] == "老保管、新媳妇来了":
        title = "老保管新媳妇来了"
    elif film["title"] == "买牛、扒瓜园":
        title = "买牛扒瓜园"
    elif film["title"] == "农家宝、两垅地":
        title = "农家宝两垅地"
        
    if not special_name:
        filmTitleToObj[title] = film
    elif fKey == film["key"]:
        filmTitleToObj[title] = film
        
for film in films:
    title = film["title"]
    key = film["key"]
    cat = film["category"]
    if len(cat) <= 0:
    	print(f"{title} ({key}): {cat}")
    
    if "、" in title:
    	print(title)
```

```python
def extractImages(pathIn, pathOut="./thumbnails/cache", thumbnail_scale=0.25):
	video_name = pathIn.split('/')[-1]
	if video_name.endswith('.mp4'):
		video_name = video_name[:-4]
	regions = filmTitleToObj[video_name]["category"]
	
	reg = ""
	if len(regions) == 1:
		reg = regions[0]
	elif len(regions) > 1:
		if regions[0] == "Canton" and regions[1] == "Hong Kong":
			reg = "Canton"
		elif regions[0] == "Hong Kong" and regions[1] == "Canton":
			reg = "Canton"
		else:
			print("Unexpected region: ", regions)
			return
	else:
		print("Unexpected no region: ")
		return

		
	parent_dir = os.path.join(pathOut, reg)
	if not os.path.exists(parent_dir):
		os.mkdir(parent_dir)

	count = 0
	vidcap = cv2.VideoCapture(pathIn)
	success,image = vidcap.read()
	success = True
	total_secs = vidcap.get(cv2.CAP_PROP_FRAME_COUNT) / vidcap.get(cv2.CAP_PROP_FPS)

	while success:
		pos = count*TIME_INTERVAL # seconds
		if pos > total_secs:
			break

		target = parent_dir + f"/{video_name}-thumb{count}.jpg"
		if os.path.exists(target):
			break
		
		vidcap.set(cv2.CAP_PROP_POS_MSEC, pos * 1000)
		success,image = vidcap.read()
		
		if success:
			image = cv2.resize(image, (0, 0), fx=thumbnail_scale, fy=thumbnail_scale)
			cv2.imwrite(filename=target, img=image)
			#free the memory used by the image
			del image
			count = count + 1
        
	vidcap.release()
	now = datetime.now()
	print(now.strftime("%Y-%m-%d %H:%M:%S"), f"Extracted {count} frames from {video_name}.")

for video in glob.glob("../../Temp/*.mp4"):
	extractImages(video, "./thumbnails/cache", SCALING_FACTOR)
```
